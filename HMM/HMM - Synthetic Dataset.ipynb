```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
N = 10  # Number of patients
D = 1   # Dimensionality of observation
T = 10  # Number of time-points
K = 2   # Number of states
```

```python
A = np.array([[0.6,0.4],[0.25,0.75]])
pi = np.random.dirichlet((3,3))
mean = np.array([0,2])
variance = np.array([0.1,0.1])
b = 2

X = np.zeros((N,T,D))
Z = np.zeros((N,T))
```

```python
for patient in range(N):
    for time in range(T):
        if time==0:
            state = np.argmax(np.random.multinomial(K,pi,size=1))
            Z[patient,time] = state
            X[patient,time,:] = np.random.normal(mean[state],variance[state],1)
        else:
            previous_state = int(Z[patient,time-1])
            previous_state_transition = A[previous_state]
            state = np.argmax(np.random.multinomial(K,previous_state_transition,size=1))
            Z[patient,time] = state
            X[patient,time,:] = np.random.normal(mean[state],variance[state],1)
```
